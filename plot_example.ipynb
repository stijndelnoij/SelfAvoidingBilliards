```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5*np.sqrt(3)]
%matplotlib widget
```

```python
d_tri = np.loadtxt("data/distance_triangle.txt")
d_sq = np.loadtxt("data/distance_square.txt")

H_tri, bins_tri = np.histogram(d_tri, bins=100)
H_sq, bins_sq = np.histogram(d_sq, bins=100)

plt.figure(1, figsize=[15,15])
plt.plot((bins_tri[1:] + bins_tri[:-1])/2., H_tri, "blue", label="triangle")
plt.plot((bins_sq[1:] + bins_sq[:-1])/2., H_sq, "red", label="square")
plt.legend()
```

```python
path = np.loadtxt("data/path.txt")
print(path.shape)
sides = 3
theta0 = 0
outer_X = np.cos(np.linspace(theta0, 2*np.pi+theta0, sides+1))
outer_Y = np.sin(np.linspace(theta0, 2*np.pi+theta0, sides+1))
M_rotation = [[np.cos(theta0), -np.sin(theta0)], [np.sin(theta0), np.cos(theta0)]]

X = path[:,0]
Y = path[:,1]
[X, Y] = np.dot(M_rotation, [X,Y])

print(np.sum(np.sqrt((X[1:]-X[:-1])**2+(Y[1:]-Y[:-1])**2))/np.sqrt(3))

plt.clf()
plt.figure(1, figsize=[15,15])

plt.plot(outer_X, outer_Y,'black')
plt.plot(X,Y, 'b')
plt.plot(X[0],Y[0],'bo')
plt.plot(X[-1],Y[-1],'ro')
```
